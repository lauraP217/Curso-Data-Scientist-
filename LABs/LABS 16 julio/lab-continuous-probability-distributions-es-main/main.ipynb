    * bottom=10, ceiling=15, count=100

El resultado debe ser como el siguiente:



```python
def genUniform(bottom,ceiling,count):
    nums = np.random.uniform(bottom,ceiling,count)
    return nums
```

```python
nums1 = genUniform(10, 15, 100)
nums1[0:20]
```

```python
nums2= genUniform(10, 60, 1000)
nums2[0:20]
```

```python
fig,ax=plt.subplots(1,2)
ax[0].hist(nums1,bins=10)
ax[1].hist(nums2,bins=10)
ax[0].set_ylim(0,115)
ax[0].set_xlim(10,15)
ax[1].set_ylim(0,115)
ax[1].set_xlim(10,60)

plt.show()
```

```python
def genNormal(mean,std_dev, numsize):
    nums = np.random.normal(loc=mean, scale=std_dev,size=numsize)
    return nums
```

```python
nums1=genNormal(10,1,1000)
nums1[0:20]
```

```python
nums2=genNormal(10,50,1000)
nums2[0:20]
```

```python
fig,ax=plt.subplots(1,2)
ax[0].hist(nums1,bins=50)
ax[1].hist(nums2,bins=50)

plt.show()
```

## Distribución Exponencial

1. Utilizando `numpy.random.exponential`, crea una función que devuelva una lista de números distribuidos exponencialmente con la media de 10. 

1. Utiliza la función para generar dos secuencias de números de tamaño 10 y 100.

1. 2. Traza las distribuciones como histogramas con el número de bins como 100.

El resultado debe ser como el siguiente:



```python
# random.exponential(scale=1.0, size=None)
def numExponencial(scale, size):
    nums = np.random.exponential(scale=scale, size=size)
    return nums
```

```python
nums1 = numExponencial(1,1000)
nums1[0:20]
```

```python
nums2 = numExponencial(100,1000)
nums2[0:20]
```

```python
fig,ax=plt.subplots(1,2)
ax[0].hist(nums1,bins=100)
ax[1].hist(nums2,bins=100)

plt.show()
```

¿En qué se diferencian ambas distribuciones?

La media cambia, por lo que la distribución también cambia.

## Distribución Exponencial de datos reales

Supongamos que el tiempo que uno pasa en un banco se distribuye exponencialmente con media de 10 minutos (es decir, λ = 1/10). ¿Cuál es la probabilidad de que un cliente pase menos de quince minutos en el banco? 

Escribe un código en python para resolver este problema

```python
#𝜆 = 1/10
tasa=1/10
tasa
```

```python
#𝑃(x < 15) = 1 − 𝑒^(−𝜆*15) 
P_menos15 = 1- math.e **(tasa*-1*15)
print(f'La probabilidad de que un cliente pase menos de quince minutos en el banco es {P_menos15*100:.2f}%')
```
